```python
dfTest = pd.read_csv("vgsales.csv", decimal = ".")
dfTest
```

```python
dfTest.info()
```

```python
dfTest.isnull().sum()
```

```python
df = dfTest.copy()
```

```python
df.dropna(axis = 0, inplace = True)
```

```python
df
```

```python
# Questão 1: Qual gênero de jogo (Genre) é o mais comum na lista?
df["Genre"].value_counts()
```

```python
# Resposta: Plataforma
```

```python
# Questão 2: Qual plataforma (Platform) teve o maior total de vendas globais (Global_Sales)?
df_maiorVenda = df.groupby("Platform")[["Global_Sales"]].sum()
df_maiorVenda.sort_values(by = "Global_Sales", ascending = False)
```

```python
# Resposta: Plataforma Wii
```

```python
# Pergunta 3: Quais são as 5 publicadoras (Publisher) com a maior média de vendas por jogo?
```

```python
df
```

```python
df_publicadora = df.groupby("Publisher")[["Global_Sales"]].mean()
```

```python
df_publicadora.sort_values(by = "Global_Sales", ascending = False)
```

```python
# Resposta: Nintendo
```

```python
# Pergunta 4: Quantos jogos do gênero 'Action' foram lançados a partir do ano 2010 (inclusive)?
```

```python
df_Action = df[(df["Genre"] == 'Action') & (df["Year"] >= 2010)]
```

```python
df_Action["Genre"].value_counts()
```
